```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("data/canada_per_capita_income.csv")
df.tail()
```

```python
plt.scatter(df["year"], df["per capita income (US$)"], color="indigo")

plt.title("Per Capita Income Over Years", fontdict={'fontsize': 16})
plt.xlabel("Year", fontdict={"fontsize": 14})
plt.ylabel("Per Capita Income (US$)", fontdict={'fontsize': 14})

plt.show()
```

```python
model = LinearRegression()
```

```python
X = df[["year"]]
```

```python
y = df["per capita income (US$)"]
```

```python
model.fit(X, y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y_predict = model.predict(X)
y_predict
```

```python
plt.scatter(df["year"], df["per capita income (US$)"], color="indigo")
plt.plot(X, y_predict, color="red")
plt.title("Per Capita Income Over Years", fontdict={'fontsize': 16})
plt.xlabel("Year", fontdict={"fontsize": 14})
plt.ylabel("Per Capita Income (US$)", fontdict={'fontsize': 14})

plt.show()
```

```python
year_to_predict = pd.DataFrame({"year": [2020]})
model.predict(year_to_predict)
```
