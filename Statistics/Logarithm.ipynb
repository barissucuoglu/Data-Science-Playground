```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.DataFrame({
    'Company': ['Tesla', 'UBER', 'Amazon', 'Jindal Steel', 'Axis Bank', 'Vedanta'],
    'Revenue (Billion $)': [31, 11, 386, 4.7, 5.6, 11.3]
})
df
```

```python
df.plot(x="Company", y="Revenue (Billion $)", kind="bar")
```

```python
df.plot(x="Company", y="Revenue (Billion $)", kind="bar", logy=True)
```

```python
df = pd.read_csv("data/income.csv",
                index_col=None,
                names=["income", "count"],
                skiprows=1)

df.head()
```

```python
sns.set(rc={"figure.figsize": (11.7, 8.27)})
g = sns.barplot(x="income", y="count", data=df)
g.set_xticklabels(g.get_xticklabels(),
                 rotation=45,
                 horizontalalignment="right");
```

```python
sns.set(rc={"figure.figsize": (11.7, 8.27)})
g = sns.barplot(x="income", y="count", data=df)
g.set_xticklabels(g.get_xticklabels(),
                 rotation=45,
                 horizontalalignment="right");
g.set(xscale="log");
```
