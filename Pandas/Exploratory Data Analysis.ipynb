# Questions

```python
# Difference in speed for the 50k, 50mi male to female
df3.groupby(["race_length", "athlete_gender"])["athlete_average_speed"].mean()
```

```python
# what age groups are the best in the 50 km Race (20 + races min)
df3.query("race_length == '50mi'").groupby("athlete_age")["athlete_average_speed"] \
.agg(["mean", "count"]).sort_values("mean", ascending=False).query("count > 19").head(15)
```

```python
# what age groups are the worst in the 50 km Race (10 + races min)
df3.query("race_length == '50mi'").groupby("athlete_age")["athlete_average_speed"] \
.agg(["mean", "count"]).sort_values("mean", ascending=True).query("count > 9").head(10)
```

```python
# Season for the data -> Slower in summer than winter?

#spring 3-5
#summer 6-8
#fall 9-11
#winter 12-2

#split between two decimals
```

```python
df3["race_month"] = df3["race_day"].str.split(".").str.get(1).astype(int)
```

```python
df3["race_season"] = df3["race_month"] \
.apply(lambda x: "Winter" if x > 11 else "Fall" if x > 6 else "Summer" if x > 5 else "Spring" if x > 2 else "Winter")
```

```python
df3.head()
```

```python
df3.groupby("race_season")["athlete_average_speed"].agg(["mean", "count"]).sort_values("mean", ascending=False)
```

```python
# 50 miler only
df3.query("race_length == '50mi'").groupby("race_season")["athlete_average_speed"] \
.agg(["mean", "count"]).sort_values("mean", ascending=False)
```
