## Contract

```python
fifa.Contract.dtype
```

```python
fifa.Contract.unique()
```

```python
for index, row in fifa.iterrows():
    if "On Loan" in row["Contract"] or "Free" in row["Contract"]:
        print(row["Contract"])
```

```python
def exract_contract_info(contract):
    if contract == "Free" or "On Loan" in contract:
        start_date = np.nan
        end_date = np.nan
        contract_length = 0
    else:
        start_date, end_date = contract.split(" ~ ")
        start_year = int(start_date[:4])
        end_year = int(end_date[:4])
        contract_length = end_year - start_year
    return start_date, end_date, contract_length
```

```python
new_cols = ["Contract Start", "Contract End", "Contract Length(years)"]
new_data = fifa.Contract.apply(lambda x: pd.Series(exract_contract_info(x)))

for i in range(len(new_cols)):
    fifa.insert(loc=fifa.columns.get_loc("Contract")+1+i, column=new_cols[i], value=new_data[i])
```

```python
fifa.sample(3)
```

```python
def categorize_contract_status(contract):
    if contract == "Free":
        return "Free"
    elif "On Loan" in contract:
        return "On Loan"
    else:
        return "Contract"

fifa.insert(loc=fifa.columns.get_loc("Contract Length(years)")+1, column="Contract Status", 
            value=fifa["Contract"].apply(categorize_contract_status))
fifa.sample(3)
```

## Heights

```python
fifa.Height.dtype
```

```python
fifa.Height.unique()
```

```python
def convert_heights(height):
    if "cm" in height:
        return int(height.strip("cm"))
    else:
        feet, inches = height.split("'")
        total_inches = int(feet) * 12 + int(inches.strip('"'))
    return round(total_inches * 2.54)

fifa.Height = fifa.Height.apply(convert_heights)
fifa.Height.unique()
```

```python
fifa = fifa.rename(columns={"Height": "Height(cm)"})
fifa.sample(2)
```

```python
fifa.Weight.dtype
```

```python
fifa.Weight.unique()
```

```python
def convert_weights(weight):
    if "kg" in weight:
        return int(weight.strip("kg"))
    else:
        pounds = int(weight.strip("lbs"))
        return round(pounds / 2.205)

fifa.Weight = fifa.Weight.apply(convert_weights)
fifa.Weight.unique()
```

```python
fifa = fifa.rename(columns={"Weight": "Weight(kg)"})
fifa.sample(2)
```
