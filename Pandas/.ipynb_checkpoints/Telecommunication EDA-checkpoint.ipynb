# Correlation and Regression Analysis

```python
correlation = df[['Bill Amount (TL)', 'Customer Satisfaction']].corr().iloc[0,1]
formatted_correlation = "{:.3f}".format(correlation)
print("Pearson Korelasyon Katsays:", correlation)
```

```python
plt.figure(figsize=(10, 6))

sns.scatterplot(x='Bill Amount (TL)', y='Customer Satisfaction', hue='City', data=df)
plt.title(f'Fatura Tutar ve Mteri Memnuniyeti Arasndaki liki - Korelasyon: {formatted_correlation}', fontsize=16)
plt.legend(title='City', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
X = df['Bill Amount (TL)']
y = df['Customer Satisfaction']
```

```python
X = np.array(X).reshape(-1, 1)

model = LinearRegression()
model.fit(X, y)

y_pred = model.predict(X)

plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', label='Veriler')
plt.plot(X, y_pred, color='red', linewidth=2, label='Regresyon Hatt')
plt.title('Ortalama Fatura Tutar ile Mteri Memnuniyeti Arasndaki liki')
plt.xlabel('Bill Amount (TL)')
plt.ylabel('Customer Satisfaction')
plt.legend()
plt.show()
```

```python
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)

print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")
```

## Customer Segmentetion

```python
# Her mteri iin ortalama fatura tutarn hesaplama
avg_bill_per_customer = df.groupby('Customer ID')['Bill Amount (TL)'].mean()

# Mterileri segmentlere ayrma (dk, orta, yksek)
avg_bill_per_customer = pd.DataFrame(avg_bill_per_customer)
avg_bill_per_customer['Segment'] = pd.qcut(avg_bill_per_customer['Bill Amount (TL)'], q=3, labels=['Dk', 'Orta', 'Yksek'])

# Segmentlerle mteri memnuniyetini birletirme
avg_satisfaction_per_customer = df.groupby('Customer ID')['Customer Satisfaction'].mean()
avg_bill_per_customer['Avg Satisfaction'] = avg_satisfaction_per_customer

avg_bill_per_customer.head()
```

```python
# Segmentlere gre ortalama mteri memnuniyeti
segment_satisfaction = avg_bill_per_customer.groupby('Segment', observed=True)['Avg Satisfaction'].mean()
segment_satisfaction
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=segment_satisfaction.index, y=segment_satisfaction.values, palette='pastel')
plt.title('Segmentlere Gre Ortalama Mteri Memnuniyeti')
plt.xlabel('Segment')
plt.ylabel('Ortalama Mteri Memnuniyeti')
plt.show()
```

```python
# ANOVA iin segmentlere ayrlm mteri memnuniyetini ekelim
low_segment = avg_bill_per_customer[avg_bill_per_customer['Segment'] == 'Dk']['Avg Satisfaction']
mid_segment = avg_bill_per_customer[avg_bill_per_customer['Segment'] == 'Orta']['Avg Satisfaction']
high_segment = avg_bill_per_customer[avg_bill_per_customer['Segment'] == 'Yksek']['Avg Satisfaction']

# ANOVA testi uygulama
f_statistic, p_value = stats.f_oneway(low_segment, mid_segment, high_segment)

print(f"ANOVA Testi F-statistii: {f_statistic:.3f}")
print(f"ANOVA Testi P-Deeri: {p_value:.3f}")
```
